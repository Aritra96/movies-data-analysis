
### Data Cleaning

#### 1. Treating null/missing values and dropping irrelevant columns

> One of the first observations is the presence of missing values or null values or NA values. This appears to be the case for a number of our columns, therefore, this section looks at treating them either by dropping wherever applicable or by imputing the missing the values.

> Dropping missing values based on low number of missing values:
For the <b>cast, director, overview and genres</b> columns, we just drop the rows with missing values since the no. of missing values are pretty low compared to the total no. of rows in this dataset.

```python
data.dropna(axis = 0, subset = ['cast', 'director', 'overview', 'genres'], inplace = True)
data.isnull().sum()
```

> We can also observe that the <b>imdb_id</b> column has some missing values but it is not exactly the primary key for this dataset. Therefore, we drop the entire imdb_id column. Additionally, since we have no use for certain other columns for the purpose of this analysis, we will omit them from the dataframe. These columns are: <b>homepage, director, tagline, cast, overview, production_companies, keywords, genres</b>

```python
data.drop(['imdb_id', 'homepage', 'director', 'tagline', 'cast', 'overview', 'production_companies', 'keywords', 'genres'], axis = 1, inplace = True)
```

```python
data.isnull().sum()
```

```python
data.info()
```

> As we can see now, we have taken care of null values and have also removed unnecessary columns (although perhaps it would have been better to perform unnecessary column removal prior to treating null values). 

> Now, we have a dataframe with <b>10730</b> observations as opposed to the original 10866 which is fair enough since we'll not be missing out on much.

#### 2. Treating 0 values (0.000000 or 0.000000e+00)

> If you remember, when we last checked for the summary statistics, we found out that there were some grave anomalies such as a <b>minimum budget or revenue of 0.000000e+00</b> values recurring for the 25th as well as the 50th percentile. This section focuses on handling those values. 



```python
data.describe()
```

> It would be prudent to view all the 0 budget and 0 revenue movies before proceeding with further analysis to find out the number of these movies since a really small number of these occurrences can be omitted from the data frame. However, if there are a lot of such observations, then we have to make a judgment call regarding how we deal with them.

```python
# Viewing all 0 budget films
zero_budget_movies = data['budget'] == 0
data[zero_budget_movies]
```

> There are about ~5600 rows with missing budget (and missing revenue as well) values given that such a high occurrence can only mean not inputing the correct data at the source of data collection. We can look manually for these values for every observation on the internet and input them ourselves but that is going to take a long time. Alternaively, we can drop these rows but that will lead to a loss of about half our dataset. Replacing them (imputing) with the mean or median would logically lead to spurious results.

> In the absence of a better alternative, we simply replace the 0's with 'null' values for the budget column. We perform the same test for the revenue column and repeat this step if the no. of 0 revenue columns are as high.

```python
# Viewing all 0 revenue movies
zero_revenue_movies = data['revenue'] == 0
data[zero_revenue_movies]
```

> The 0 revenue movies are even more than the 0 budget movies indicating that a major chunk of our dataset may be compromised should be choose to drop them.

> Our intended course of action, given the circumstances, would be to replace 0 with null values for the budget and revenue columns.


```python
# Viewing all 0 runtime movies
zero_runtime_movies = data['runtime'] == 0
data[zero_runtime_movies]
```

```python
# Replacing 0 values with null for budget, runtime, and revenue columns using numpy
data['budget'] = data['budget'].replace(0, np.NaN)
data['revenue'] = data['revenue'].replace(0, np.NaN)
data['runtime'] = data['runtime'].replace(0, np.NaN)

# Checking the changes in the data
data.info()
```

> We now have a dataset with more null than non-null values for budget, runtime and revenue columns. Let us try repeating this process for budget_adj and revenue_adj columns and take the necessary actions.

```python
data['budget_adj'] = data['budget_adj'].replace(0, np.NaN)
data['revenue_adj'] = data['revenue_adj'].replace(0, np.NaN)

data.info()
```

> At this juncture, it is important to address the budget vs budget_adj column and the revenue vs revenue_adj column. Let us try to find if these columns are in fact substitutable. 

```python
# Checking the correlation coefficient between the budget and the budget_adj columns
data['budget'].corr(data['budget_adj'])
```

```python
# Checking the correlation coefficient between the revenue and the revenue_adj columns
data['revenue'].corr(data['revenue_adj'])
```

> As it turns out, there is a strong positive correlation between the aforementioned columns. Therefore, we act on our presupposition of substitutes. As a result, we proceed to drop the budget and the revenue columns.


```python
data.drop(['budget', 'revenue'], axis = 1, inplace = True)
```

```python
# Viewing the columns once more
data.info()
```

### Research Question 2 - Does the release date (month) of a movie/tv series affect its popularity, revenue, vote count, and vote average?

> This question explores the relationship between the release month of a movie and its popularity, revenue, vote count and vote average. The idea here is to argue whether a particular month of release may be best suited in order to, say, bag the most revenue or ensure that a movie enjoys high popularity.

#### Data Visualization and Analysis

```python
# Our first order of business here is to extract a month column from the release_date column.

data['month'] = data['release_date'].apply(lambda x: x.month)
data.head()
```

```python
# Let us start with month vs revenue
# Group by month and aggregate on the revenue by sum

revenue_by_month = data.groupby('month')['revenue_adj'].sum()
revenue_by_month
```

> We notice that the 12th and the 6th months are the highest grossing months in terms of revenue, followed by the 5th and the 7th months. It may not be prudent to make the conclusion about the best months yet since the revenue is also a function of the number of movies released in those months. Therefore, we count the movies before proceeding.

```python
# Finding the count of movies per month
data['month'].value_counts()
```

```python
# Finding the mean no. of movies released per year
data['month'].value_counts().mean()
```

> We can clearly see how the number of movies released can skew the revenue. In this case, the 9th and 10th months have a lot more releases and a really low gross revenue when compared to the 6th and 12th months. 

> The 6th and 12th months have 822 and 974 releases respectively which is in the vicinity of the mean no. of movies released per year (894). Therefore, we maintain that the month of June and December are the winners here - December being the more ideal month.

> We now repeat this process for the remaining variables.

```python
# Month vs popularity
# Group by month and aggregate on the popularity by sum

pop_by_month = data.groupby('month')['popularity'].sum()
pop_by_month
```

> Here, the clear top two contenders for popularity are the months of September and December. It is interesting to note here that December emerges as a winning candidate yet again and the popularity of September has potential of being influenced since it is the month with the highest number of movie releases. 

> Accounting for all explored factors, we maintain December as the better of all months in terms of popularity.

```python
# Month vs vote count
# Group by month and aggregate on the vote count by sum

vc_by_month = data.groupby('month')['vote_count'].sum()
vc_by_month
```

> With respect to vote counts, the month of June and December reappear as the winners followed closely by May and October. 

> Having accounted for all explored factors, December stands out as one of the best contenders in terms of vote counts.

```python
# Month vs vote average
# Group by month and aggregate on the vote average by sum

va_by_month = data.groupby('month')['vote_average'].sum()
va_by_month
```

> The top two months for vote average are September and October, followed by December. 

> Accounting for all explored factors, September seems to be the clear winner here.

```python
# Plot for Month vs Revenue
data.plot(x = 'month', y = 'revenue_adj', kind = 'scatter', figsize = (10, 5));
```

```python
# Plot for Month vs Popularity
data.plot(x = 'month', y = 'popularity', kind = 'scatter', figsize = (10, 5));
```

```python
# Plot for Month vs Vote count
data.plot(x = 'month', y = 'vote_count', kind = 'scatter', figsize = (10, 5));
```

```python
# Plot for Month vs Vote average
data.plot(x = 'month', y = 'vote_average', kind = 'scatter', figsize = (10, 5));
```

> Through the visualizations we are able to graphically affirm our findings. We can also notice the presence of outliers in the graphs.

#### Inferences

> We could infer the following about the 4 pairs of variables chosen for this analysis:

> <b>1. Month and Revenue -</b> If you want to maximize your revenue, the months of June and December (December moreso) are likely to be the best months for movie release.

> <b>2. Month and Popularity -</b> If you want to maximize your popularity, the months of September and December are likely to be the best months for movie release.

> <b>3. Month and Vote count -</b> If you want to maximize your vote count, the months of June and December are likely to be the best months for movie release.

> <b>4. Months and Vote average -</b> If you want to maximize your vote average, the months of September and October are likely to be the best months for movie release.

#### Summary Report

> This analysis of the TMDb dataset has revealed interesting insight for the questions that were posed at the beginning. These questions might be helpful for production houses and movie makers given the kind of information that we have generated from this exploration.

> In summary:

> <b> Research Question 1 -</b> Revenue-popularity and revenue-vote count seem to enjoy an inverse relationship. As counterintuitive as it may seem, a higher revenue may not lead to a higher popularity and vice-versa. Additionally, a higher revenue may not yield higher vote counts and vice-versa.

> <b>Note:</b> The relationship between vote count and vote average have not been separately explored in this project.

> <b> Research Question 2 -</b> The better of all months for improving a movie's revenue, popularity, and vote count is the month of December. The month of September is positively related with improving vote average. Therefore, we can say, with some probability, that December is the best month for production houses and movie makers when it comes to release. This could be attributed to a number of factors such as festival time, increase in income, appraisals, bonuses, et al.
